```python
#Importando as bibliotecas necessárias

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Abrindo o arquivos de dado

df_2014 = pd.read_csv('/content/votacao_secao_2014_RR.csv', sep=';', encoding='ISO-8859-1')
df_2014.head()
```

```python
type(df_2014)
```

```python
df_2014.columns
```

```python
df_2014_fltrado = df_2014[['NM_MUNICIPIO',	'NR_ZONA', 'NR_SECAO', 'DS_CARGO', 'NR_VOTAVEL','NM_VOTAVEL','QT_VOTOS']]
df_2014_fltrado
```

```python
dep_estadual = df_2014_fltrado[df_2014_fltrado['DS_CARGO'] == 'Deputado Estadual']
dep_estadual
```

```python
estadual_mais_votados_2014 = dep_estadual.groupby(['NM_VOTAVEL']).agg({'QT_VOTOS': 'sum'})
estadual_mais_votados_2014
```

```python
estadual_sorted = estadual_mais_votados_2014.sort_values('QT_VOTOS', ascending=False)
estadual_sorted.head(20)
```

```python
lenir = dep_estadual[dep_estadual['NM_VOTAVEL'] == 'LENIR RODRIGUES SANTOS']
lenir 
```

```python
df_lenir_por_municipios = lenir.groupby(['NM_MUNICIPIO']).agg({'QT_VOTOS': 'sum'})
lenir_municipio = df_lenir_por_municipios.apply(lambda x: x.sort_values(ascending=False))
lenir_municipio
```
